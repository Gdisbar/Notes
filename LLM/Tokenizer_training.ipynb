```python
from datasets import load_dataset
from tokenizers import Tokenizer, models, trainers, pre_tokenizers, decoders

# Step 1: Load a dataset
dataset = load_dataset("wikitext", "wikitext-2-raw-v1", split="train")

dataset
```

```python
# Step 2: Extract text from the dataset
texts = dataset['text']

# Step 3: Initialize a tokenizer
tokenizer = Tokenizer(models.BPE())
tokenizer.pre_tokenizer = pre_tokenizers.Whitespace()

# Step 4: Define a trainer
trainer = trainers.BpeTrainer(vocab_size=30000, min_frequency=2)

```

```python
# Step 5: Train the tokenizer
tokenizer.train_from_iterator(texts, trainer=trainer)


# Step 6: Save the tokenizer
tokenizer.save("tokenizer.json")
```

```python

# Step 7: Load the tokenizer
tokenizer = Tokenizer.from_file("tokenizer.json")

# Example usage
encoded = tokenizer.encode("Hello, how are you?")
print("Encoded IDs:", encoded.ids)
tokenized = tokenizer.decode(encoded.ids)
print("Tokenized text:", tokenized)
# Iterate through the list of token IDs to get individual tokens
for token_id in encoded.ids:
    token = tokenizer.id_to_token(token_id)  # Get the token for each ID
    print(f"{token_id} : {token}")
```
